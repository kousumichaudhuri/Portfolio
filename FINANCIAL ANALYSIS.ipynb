```python
# Display the first few rows of the dataset
print(data.head())

# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Histogram for Market Capitalization
plt.hist(data['Mar Cap - Crore'], bins=20)
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Frequency')
plt.title('Distribution of Market Capitalization')
plt.show()
```

```python
# Histogram for Quarterly Sales
plt.hist(data['Sales Qtr - Crore'], bins=20)
plt.xlabel('Quarterly Sales (Crore)')
plt.ylabel('Frequency')
plt.title('Distribution of Quarterly Sales')
plt.show()
```

```python
# Calculate growth rate
data['Growth Rate'] = (data['Sales Qtr - Crore'] - data['Sales Qtr - Crore'].shift(1)) / data['Sales Qtr - Crore'].shift(1) * 100

# Calculate profitability ratio (example)
data['Profitability Ratio'] = data['Mar Cap - Crore'] / data['Sales Qtr - Crore']
```

```python
# Correlation analysis
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
# Scatter plot for Market Capitalization vs. Quarterly Sales
plt.scatter(data['Mar Cap - Crore'], data['Sales Qtr - Crore'])
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.title('Market Capitalization vs. Quarterly Sales')
plt.show()
```
