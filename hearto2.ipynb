```python
# Define a threshold for spikes (e.g., O2 Saturation > 95%)
threshold = 95
spikes = df2[df2['O2 Saturation'] > threshold]

# Plotting the O2 saturation levels with spikes highlighted
plt.figure(figsize=(10, 6))
plt.plot(df2.index, df2['O2 Saturation'], label='O2 Saturation')
plt.scatter(spikes.index, spikes['O2 Saturation'], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Spikes Highlighted')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define a specific range to zoom in
start = 0
end = 500  # Adjust these values as needed

# Plot the O2 saturation levels with spikes highlighted for the specific range
plt.figure(figsize=(10, 6))
plt.plot(df2.index[start:end], df2['O2 Saturation'][start:end], label='O2 Saturation')
plt.scatter(spikes.index[(spikes.index >= start) & (spikes.index < end)], spikes['O2 Saturation'][(spikes.index >= start) & (spikes.index < end)], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Spikes Highlighted (Zoomed In)')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define a higher threshold for spikes (e.g., O2 Saturation > 98%)
new_threshold = 98
spikes = df2[df2['O2 Saturation'] > new_threshold]

plt.figure(figsize=(10, 6))
plt.plot(df2.index, df2['O2 Saturation'], label='O2 Saturation')
plt.scatter(spikes.index, spikes['O2 Saturation'], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Higher Threshold for Spikes')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Apply a rolling average with a window size of 50
df2['O2 Saturation Rolling'] = df2['O2 Saturation'].rolling(window=50).mean()

plt.figure(figsize=(10, 6))
plt.plot(df2.index, df2['O2 Saturation Rolling'], label='O2 Saturation (Rolling Average)')
plt.scatter(spikes.index, spikes['O2 Saturation'], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Rolling Average')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define a specific range to zoom in
start = 0
end = 500  # Adjust these values as needed

plt.figure(figsize=(10, 6))
plt.plot(df2.index[start:end], df2['O2 Saturation'][start:end], label='O2 Saturation')
plt.scatter(spikes.index[(spikes.index >= start) & (spikes.index < end)], spikes['O2 Saturation'][(spikes.index >= start) & (spikes.index < end)], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Spikes Highlighted (Zoomed In)')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example baseline for O2 saturation (e.g., 95%)
baseline = 95

plt.figure(figsize=(10, 6))
plt.plot(df2.index, df2['O2 Saturation'], label='O2 Saturation')
plt.axhline(y=baseline, color='green', linestyle='--', label='Clinical Baseline')
plt.scatter(spikes.index, spikes['O2 Saturation'], color='red', label='Spikes', zorder=5)
plt.title('O2 Saturation Levels with Clinical Baseline')
plt.xlabel('Index')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming df1 and df2 are already loaded and df1 has a 'time' column
df_combined = pd.concat([df1, df2], axis=1)

# Example: Scatter plot to compare heart rate (e.g., 'thalachh') and O2 saturation
plt.figure(figsize=(10, 6))
plt.scatter(df_combined['thalachh'], df_combined['O2 Saturation'], alpha=0.5)
plt.title('Heart Rate vs O2 Saturation')
plt.xlabel('Heart Rate (thalachh)')
plt.ylabel('O2 Saturation (%)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Plot a histogram to see the distribution of O2 saturation levels
plt.figure(figsize=(10, 6))
sns.histplot(df2['O2 Saturation'], bins=30, kde=True)
plt.title('Distribution of O2 Saturation Levels')
plt.xlabel('O2 Saturation (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot a scatter plot of heart rate (thalachh) vs. O2 saturation
plt.figure(figsize=(10, 6))
plt.scatter(df_combined['thalachh'], df_combined['O2 Saturation'], alpha=0.5)
plt.title('Heart Rate vs O2 Saturation')
plt.xlabel('Heart Rate (thalachh)')
plt.ylabel('O2 Saturation (%)')
plt.grid(True)
plt.show()
```

```python
# Highlight points around 97.5% O2 saturation
within_range = df_combined[(df_combined['O2 Saturation'] >= 97) & (df_combined['O2 Saturation'] <= 98)]
out_of_range = df_combined[(df_combined['O2 Saturation'] < 97) | (df_combined['O2 Saturation'] > 98)]

plt.figure(figsize=(10, 6))
plt.scatter(within_range['thalachh'], within_range['O2 Saturation'], alpha=0.5, label='Within 97-98%', color='blue')
plt.scatter(out_of_range['thalachh'], out_of_range['O2 Saturation'], alpha=0.5, label='Out of 97-98%', color='red')
plt.title('Heart Rate vs O2 Saturation with Highlighted Ranges')
plt.xlabel('Heart Rate (thalachh)')
plt.ylabel('O2 Saturation (%)')
plt.legend()
plt.grid(True)
plt.show()
```
