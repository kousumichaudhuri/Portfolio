```python
# Fill missing values or drop rows/columns with missing values
data = data.dropna()  # Simple approach, you can use more sophisticated imputation if needed

# Convert data types if necessary
data['Age'] = data['Age'].astype(int)
data['DailyRate'] = data['DailyRate'].astype(float)

# Verify changes
print(data.dtypes)
```

```python
# Summary statistics
print(data.describe())

# Histograms for numerical columns
data.hist(bins=20, figsize=(20, 15))
plt.show()

# Count plots for categorical columns
categorical_columns = ['AgeGroup', 'Attrition', 'BusinessTravel', 'Department', 'EducationField']
for col in categorical_columns:
    sns.countplot(x=col, data=data)
    plt.show()

# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, fmt='.2f')
plt.show() 
```

```python
from scipy.stats import ttest_ind, chi2_contingency

# T-test for numerical variables between two groups (e.g., Attrition Yes/No)
attrition_yes = data[data['Attrition'] == 'Yes']['Age']
attrition_no = data[data['Attrition'] == 'No']['Age']
t_stat, p_val = ttest_ind(attrition_yes, attrition_no)
print(f'T-test for Age: t-statistic={t_stat}, p-value={p_val}')

# Chi-square test for categorical variables
contingency_table = pd.crosstab(data['Attrition'], data['Department'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-square test for Department: chi2={chi2}, p-value={p}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Encode categorical variables
le = LabelEncoder()
data['Attrition'] = le.fit_transform(data['Attrition'])

# Select features and target
features = ['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'JobSatisfaction', 'WorkLifeBalance']
target = 'Attrition'
X = data[features]
y = data[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('ROC AUC Score:', roc_auc_score(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Encode categorical variables
le = LabelEncoder()
data['Attrition'] = le.fit_transform(data['Attrition'])

# Select features and target
features = ['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'JobSatisfaction', 'WorkLifeBalance']
target = 'Attrition'
X = data[features]
y = data[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply SMOTE to the training data
sm = SMOTE(random_state=42)
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train)

# Build the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_smote, y_train_smote)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print('ROC AUC Score:', roc_auc_score(y_test, y_pred))
```
