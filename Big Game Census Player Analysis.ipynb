```python
# 1. Analyze player birthplace information
player_origins = df3["Geography, full name (City, State)"]

# Split player birthplaces into city and state
player_origins_split = player_origins.str.split(", ", expand=True)
player_origins_split.columns = ["City", "State"]

# Calculate the distribution of players' birthplaces by state
player_origins_distribution = player_origins_split["State"].value_counts()

# Display the distribution of players' birthplaces
print("Distribution of Players' Birthplaces by State:")
print(player_origins_distribution)
```

```python
# Convert the "Years Played" column to numeric
df1["Years Played"] = pd.to_numeric(df1["Years Played"], errors="coerce")

# Calculate interesting statistics such as the average age, weight, and years played of the players
average_age = df1["Player Age"].mean()
average_weight = df1["Player Weight (lbs.)"].mean()
average_years_played = df1["Years Played"].mean()

print("\nAverage Player Age:", average_age)
print("Average Player Weight:", average_weight)
print("Average Years Played:", average_years_played)
```

```python
# Analyze the distribution of player birthplaces
player_birthplace_counts = df1["Player Birth State"].value_counts()

# Display the distribution of player birthplaces
print("Distribution of Player Birthplaces:")
print(player_birthplace_counts)

# Visualize the distribution of player birthplaces (optional)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
player_birthplace_counts.plot(kind='bar')
plt.title('Distribution of Player Birthplaces')
plt.xlabel('State')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the distribution of player positions
player_position_counts = df1["Player Position"].value_counts()

# Display the distribution of player positions
print("\nDistribution of Player Positions:")
print(player_position_counts)

# Visualize the distribution of player positions (optional)
plt.figure(figsize=(10, 6))
player_position_counts.plot(kind='bar')
plt.title('Distribution of Player Positions')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the top player colleges
top_colleges = df1["Player College"].value_counts().head(10)

# Display the top player colleges
print("\nTop Player Colleges:")
print(top_colleges)

# Visualize the top player colleges (optional)
plt.figure(figsize=(10, 6))
top_colleges.plot(kind='bar')
plt.title('Top Player Colleges')
plt.xlabel('College')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the distribution of player ages
plt.figure(figsize=(10, 6))
plt.hist(df1["Player Age"], bins=20, color='skyblue', edgecolor='black')
plt.title('Player Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the distribution of player weights
plt.figure(figsize=(10, 6))
plt.hist(df1["Player Weight (lbs.)"], bins=20, color='salmon', edgecolor='black')
plt.title('Player Weight Distribution')
plt.xlabel('Weight (lbs)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the distribution of years played by players
plt.figure(figsize=(10, 6))
plt.hist(df1["Years Played"], bins=15, color='lightgreen', edgecolor='black')
plt.title('Player Years Played Distribution')
plt.xlabel('Years Played')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Analyze the distribution of player teams across conferences
team_conference_counts = df1["Conference"].value_counts()

# Display the distribution of player teams across conferences
print("\nDistribution of Player Teams Across Conferences:")
print(team_conference_counts)

# Visualize the distribution of player teams across conferences (optional)
plt.figure(figsize=(10, 6))
team_conference_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Player Teams Across Conferences')
plt.xlabel('Conference')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
